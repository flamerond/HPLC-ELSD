### 1. Background

To determine MWCO curves of NF and UF membranes, an HPLC-ELSD method was developed together with the Separation Science Group (Prof. Lynen). The raw data obtained from the apparatus is in the form of a .csv file containing 3 columns:

1. Peak number
2. Retention time
3. Area

The goal of this script is to enable automatic data processing of the different peaks in the chromatogram and to calculate concentrations of the different peaks in a calibration curve. 

In a later stage, the peak numbers will be assigned to certain molecular weights. For this purpose, either the MW distribution has to be clear from the pure standards (PEG 200, 300, 400, 1000) or the MW can be determined with mass spectrometry.

### 2. Importing the data

```python
# data should be saved as a UTF-8 compatible csv file, data on the HPLC computer is encoded as UTF-16

import pandas as pd
import numpy as np
import glob 

# getting csv files from the specified folder
# double backslashes are used because a single backslash is used as an end statement in python

path = 'C:\\Users\\cdierend\\OneDrive - UGent\\Universiteit Gent\\Research\\FWO Biostable\\6_MWCO\\Exp_6.1\\2_RawData\\20220202_NF_mix_calibration\\'
filenames = glob.glob(path + "\*.csv")

dataframes_list = []

# for loop to iterate over all csv files in the specified folder

for file in filenames: 
    temp_df = pd.read_csv(file, encoding = "UTF-16", sep = '\t', decimal = ',', names = ["Peak number","Retention time", "Area"])
    temp_df['Retention time'] = pd.to_numeric(temp_df['Retention time'])
    temp_df['Area'] = pd.to_numeric(temp_df['Area'])
    index = temp_df.index
    index.name = file
    dataframes_list.append(temp_df)
    
    
for dataset in dataframes_list:
    display(dataset)
    
```

### 3. Exploring the data visually

First, we want to check how the areas look in function of retention time by making a plot for each of the dataframes.

```python
import matplotlib.pyplot as plt

# the plot below just visualizes the area that is reported per retention time

titles = ['200_ppm_1', '200_ppm_2', '200_ppm_3', '400_ppm_1', '400_ppm_2', '400_ppm_3', '50_ppm_1', '50_ppm_2','50_ppm_3']

for dataset,i in zip(dataframes_list,titles):
    dataset.plot.scatter(x="Retention time", y="Area")
    plt.xlabel('Retention time (min)')
    plt.ylabel('Area')
    plt.title(i)
    plt.show

```
